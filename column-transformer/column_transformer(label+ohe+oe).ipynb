## column Transformer

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
```

```python
df=pd.read_csv('covid_toy.csv')
df
```

```python
df.isnull().sum()
```

```python
df['city'].value_counts()  #to get count of each category in column
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop(columns=['has_covid']),df['has_covid'],test_size=0.2)
```

#### doing manualy

```python
#imputer to fever colmn
si=SimpleImputer()
X_train_fever=si.fit_transform(X_train[['fever']])
X_test_fever=si.transform(X_test[['fever']])
X_train_fever  #missing value get replace by mean
```

```python
#OrdinalEncoding on cough
oe=OrdinalEncoder(categories=[['Mild','Strong']])
X_train_cough=oe.fit_transform(X_train[['cough']])
X_test_cough=oe.transform(X_test[['cough']])
X_train_cough
```

```python
#onehotEncoding on gender and city
ohe=OneHotEncoder(drop='first',sparse_output=False)
X_train_gen_city=ohe.fit_transform(X_train[['gender','city']])
X_test_gen_city=ohe.transform(X_test[['gender','city']])
X_train_gen_city
```

```python
X_train_gen_city.shape

```

```python
X_train_fever.shape
```

```python
# Extracting Age
X_train_age = X_train.drop(columns=['gender','fever','cough','city']).values

# also the test data
X_test_age = X_test.drop(columns=['gender','fever','cough','city']).values

X_train_age.shape
```

```python
X_train_transformed = np.concatenate((X_train_age,X_train_fever,X_train_gen_city,X_train_cough),axis=1)
# also the test data
X_test_transformed = np.concatenate((X_test_age,X_test_fever,X_test_gen_city,X_test_cough),axis=1)

X_train_transformed.shape
```

## column transformer

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(transformers=[
    ('tnf1',SimpleImputer(),['fever']),
    ('tnf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('tnf3',OneHotEncoder(sparse_output=False,drop='first'),['gender','city'])
],remainder='passthrough')
```

```python
transformer.fit_transform(X_train).shape
```

```python
transformer.transform(X_test).shape
```
